```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("Mall_Customers.csv")

# Look at the first few rows
print(df.head())
```

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

k = 3  # number of clusters (you can adjust later)
kmeans = KMeans(n_clusters=k, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(df.iloc[:,0], df.iloc[:,1], c=df['Cluster'], cmap='viridis', s=100)

centers = scaler.inverse_transform(kmeans.cluster_centers_)
plt.scatter(centers[:,0], centers[:,1], c='red', s=300, marker='X')

plt.xlabel(X.columns[0])
plt.ylabel(X.columns[1])
plt.title("Customer Segments")
plt.show()
```

```python
df.to_csv("clustered_customers.csv", index=False)
print("✅ Results saved to clustered_customers.csv")
```
